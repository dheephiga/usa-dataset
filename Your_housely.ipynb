```python
import pandas as pd
import numpy as np
import tensorflow as tf
import sklearn as skl
```

```python
data = pd.read_csv("https://raw.githubusercontent.com/dheephiga/usa-dataset/main/train.csv")
```

```python
data.isnull().sum()
```

```python
data["Number_of_Windows"].fillna(0, inplace = True)
data.isnull().sum()
```

```python
data["Furnishing"].fillna("No data", inplace = True)
data.isnull().sum()
```

```python
data["Frequency_of_Powercuts"].fillna("No data", inplace = True)
data["Crime_Rate"].fillna("No data", inplace = True)
data["Dust_and_Noise"].fillna("No data", inplace = True)
data.isnull().sum()
```

```python
data.shape
```

```python
from sklearn.preprocessing import OneHotEncoder
encoder = OneHotEncoder(handle_unknown = 'ignore')
```

```python
encoder_data = pd.DataFrame(encoder.fit_transform(data[['Furnishing']]).toarray())
```

```python
final_d = data.join(encoder_data)
print(final_d)
```

```python
from pandas.core.groupby.grouper import final
final_d.drop('Furnishing', axis = 1, inplace = True)
print(final_d)
```
